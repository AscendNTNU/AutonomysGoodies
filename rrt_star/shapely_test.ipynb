```python
import numpy as np
import matplotlib.pyplot as plt

import shapely
from shapely import Point, Polygon, LineString, GeometryCollection
```

```python

pt1 = shapely.Point(10, 10)
pt2 = shapely.Point(20, 20)
ln = shapely.LineString([(0, 1), (20, 100), (100, 3), (120, 102), (200, 5)])
ln
```

```python
points = []
for i in range(0, 450):
    points.append(ln.interpolate(i))
GeometryCollection(points)
```

```python
poly = Polygon([(0, 0), (1, 0), (1, 1), (0, 1)])
poly
```

```python
poly.contains(Point(0.7, 0.7))
```

```python
test_line = LineString([(-3, -3), (0.5, 0.5)])
intersections = poly.intersection(test_line)
intersections.xy
```
